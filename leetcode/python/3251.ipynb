## range of current value

from position n-1 to 0, we determine the number of arrs we can have for each postion

dp[i][j] means, for nums[i:], when the arr1 value at i is j, the number arrays. \
i.e.  for current postion i, the value is from [0, nums[i]], if we choose value j, then we can only choose values >= j at i+1 position. \
but, we only garuantee the arr1, we also need to consider arr2 \
for nums[i] and nums[j] \
we chose cur value from [0, nums[i]], that is   0 <= arr1[i] <= nums[i], then   arr2[i] = nums[i] - arr1[i] \
for i+1,  arr1[i+1] should choose from [nums[i], inf]  then     arr2[i+1] = nums[i+1] - arr1[i+1] should meet arr2[i+1] <= arr2[i] \
arr2[i+1] <= arr2[i] =>  nums[i+1] - arr1[i+1] <= nums[i] - arr1[i] => arr1[i+1] >= nums[i+1] - nums[i] + arr1[i] \
also, arr1[i+1] >= arr1[i] \
=> arr1[i+1] >= max(arr1[i], nums[i+1] - nums[i] + arr1[i])

```python
from utils.show import show2D

def countOfPairs(nums):
    n = len(nums)
    # dp = [collections.defaultdict(int) for _ in range(n)] 
    dp = [[0] * 1001 for _ in range(n)]

    # init 
    for j in range(nums[n-1]+1):
        dp[n-1][j] = 1

    show2D(dp)

    for i in range(n-2, -1, -1):
        delta = max(0, nums[i+1] - nums[i])
        print(i, delta)
        dp[i][nums[i]] = sum(dp[i+1][nums[i]+delta:])
        for j in range(nums[i]-1, -1, -1):
            dp[i][j] = dp[i][j+1] + dp[i+1][j+delta]


    show2D(dp)

    
nums = [2, 3, 2]
countOfPairs(nums)
```

```python
import collections
from utils.show import show2D

def countOfPairs(nums):
    n = len(nums)
    # dp = [collections.defaultdict(int) for _ in range(n)] 
    dp = [[0] * 1001 for _ in range(n)]

    # init 
    for j in range(nums[n-1]+1):
        dp[n-1][j] = 1

    show2D(dp)

    for i in range(n-2, -1, -1):
        for j in range(nums[i]+1):
            start = max(j + nums[i+1]-nums[i], j)
            dp[i][j] = sum(dp[i+1][start:])


    show2D(dp)

    
nums = [2, 3, 2]
countOfPairs(nums)
```
