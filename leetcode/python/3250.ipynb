```python
from utils.show import show2D

def countOfPairs(nums):
    n = len(nums)
    dp = [[0] * 51 for _ in range(n)]

    # init 
    for j in range(nums[n-1]+1):
        dp[n-1][j] = 1

    for i in range(n-2, -1, -1):
        for j in range(nums[i]+1):
            start = max(j + nums[i+1]-nums[i], j)
            dp[i][j] = sum(dp[i+1][start:]) % (10**9 + 7)
    
    show2D(dp)
    
    return sum(dp[0]) % (10**9 + 7)


nums = [2,3,2]
# nums = [23,26,29,30,35,40,42,42,43,50]
# nums = [5,5,5,5]
countOfPairs(nums)
```

```python
from functools import cache

def countOfPairs(nums):
    n = len(nums)

    @cache
    def dfs(i, last1):
        if i == n:
            return 1
        last2 = nums[i-1] - last1
        start = max(last1, nums[i]-last2)

        tmp = 0
        for val in range(start, nums[i]+1):
            tmp += dfs(i+1, val)
        print(i, last1, last2, start, tmp)
        return tmp
    
    # init 
    ans = 0
    for v in range(nums[0]+1):
        ans += dfs(1, v)
    return ans 

nums = [23,26,29,30,35,40,42,42,43,50]
# nums = [2,3,2]
countOfPairs(nums)

```
